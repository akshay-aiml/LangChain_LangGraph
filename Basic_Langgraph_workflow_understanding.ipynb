```python
from langgraph.graph import StateGraph,START,END
from typing import TypedDict
```

```python
# define the State
class SpeedState(TypedDict):
    speed:int
    distance:float
    time: float
    categori : str
```

```python
# speed formula for node

def calculate_speed(State:SpeedState) -> SpeedState :
   speed = State['distance']/State['time']
   State["speed"] = round(speed,2)
   return State
```

```python
#  categarize the speed
def categorize_speed(State:SpeedState) -> SpeedState :
    if State["speed"] < 20:
      category = "slow"
    elif State["speed"] < 60:
      category = "medium"
    else:
      category = "fast"
    State["categori"] = category
    return State
```

```python
# create graph
graph = StateGraph(SpeedState)

# all nodes
graph.add_node("SpeedCal",calculate_speed)
graph.add_node("Categorize",categorize_speed)

#add Edges
graph.add_edge(START,"SpeedCal")
graph.add_edge("SpeedCal","Categorize")
graph.add_edge("Categorize",END)
```

```python
#complile the graph
workflow = graph.compile()
```

```python
# Execute the graph
initial_state = {"distance": 360, "time":3}

result = workflow.invoke(initial_state)
print(result)
```
