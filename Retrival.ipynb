Retrieval is the process of searching and selecting the most relevant documents or data for a user query.

Types of Retrieval Techniques:
```
Retrieval Techniques
```

#Sparse Retrieval (Keyword Search)

**Sparse retrieval works using exact keyword matching.**

It is used in traditional search engines.

Examples:

BM25

TF-IDF

###The industry standard algorithm for this is BM25 (Best Matching 25).


````python
!pip install -q  rank_bm25 langchain-community langgraph  langchain-groq
````

````python
from langchain_community.retrievers import BM25Retriever
from langchain_core.documents import Document
````

````python
docs = [
    Document(page_content="The model SKU-99X uses a specialized lithium-ion battery."),
    Document(page_content="Standard lithium batteries are common in most consumer electronics."),
    Document(page_content="How to troubleshoot Error-404 on the web portal."),
    Document(page_content="Policy for PTO (Paid Time Off) for full-time employees.")
]
````

````python
# Initialize the BM25 Retriever
# This calculates the statistics (TF-IDF) for your specific corpus
retriever = BM25Retriever.from_documents(docs)
````

````python
# Set how many results to return
retriever.k = 2
````

````python
# Search for a specific technical term
query = "What battery does SKU-99X use?"
result = retriever.invoke(query)
````

````python
for doc in result:
    print(f"Match found: {doc.page_content}")
````

#Dense Retrieval (Vector Search)

**Dense retrieval uses embeddings to understand the semantic meaning of text.**

Instead of keywords, it compares vectors.

Dense Retrieval converts text into a series of numbers called Embeddings using a Neural Network (like BERT or OpenAI's text-embedding-3-small).

Dense: Unlike Sparse vectors (mostly zeros), these vectors are "dense" with floating-point numbers.

````python
from langchain.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
````


````python
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
````

````python
docs = [
    Document(page_content="The restaurant serves excellent poultry dishes."),
    Document(page_content="I am looking for a place that has great chicken."),
    Document(page_content="The weather in London is quite rainy today."),
    Document(page_content="New York is known as the Big Apple.")
]
````

````python
# Create the Vector Store (Dense Index)
# This converts all docs into vectors and indexes them
vectorstore = FAISS.from_documents(docs, embedding_model)
````

````python
# Perform a Semantic Search
# Note: The query uses "bird" and "food", words not in the first doc!
query = "Where can I find some bird based food?"
results = vectorstore.similarity_search(query, k=1)
````

````python
print(f"Top Semantic Match: {results[0].page_content}")
````

#Hybrid Retrieval (Industry Standard)

**Hybrid retrieval combines: Sparse Retrieval + Dense Retrieval**



````python
from langchain_community.retrievers import BM25Retriever
from langchain_community.vectorstores import FAISS
from langchain.retrievers import EnsembleRetriever
from langchain.embeddings import HuggingFaceEmbeddings

````

````python
texts = [
    "The iPhone 15 Pro uses a Titanium frame for better durability.",
    "Apple's latest mobile device features a high-performance A17 chip.",
    "The new smartphone from Samsung, the S24, has AI translation features.",
    "Titanium is a chemical element with the symbol Ti and atomic number 22."
]
````

````python
# Setup Sparse Retriever (BM25)
# Great for: "iPhone 15", "Titanium"
bm25_retriever = BM25Retriever.from_texts(texts)
bm25_retriever.k = 2
````

````python
# 3. Setup Dense Retriever (FAISS)
# Great for: "latest Apple phone", "durable materials"
embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

faiss_vectorstore = FAISS.from_texts(texts, embedding_model)
faiss_retriever = faiss_vectorstore.as_retriever(search_kwargs={"k": 2})
````

````python
# Create the Ensemble (Hybrid) Retriever
# We give 50% weight to each. If one is consistently better for your data, you can adjust to [0.3, 0.7]

hybrid_retriever = EnsembleRetriever(
    retrievers=[bm25_retriever, faiss_retriever],
    weights=[0.5, 0.5]
)
````

````python
query = "Tell me about the durable Apple phone"
docs = hybrid_retriever.invoke(query)
````

````python
for i, doc in enumerate(docs):
    print(f"Rank {i+1}: {doc.page_content}")
````

The core conclusion from the Hybrid Retrieval output is that neither keyword search (BM25) nor semantic search (FAISS) is perfect on its own, but together they cover each other's blind spots.
