```python
from google.colab import userdata
from langchain_groq import ChatGroq
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
import os
```

```python
api_key = userdata.get("GROQ_API_KEY")
# Initialize LLM

llm = ChatGroq(
    api_key=api_key,
    model="llama-3.3-70b-versatile",
    temperature=0.2
)
```

```python
### testing the set up
prompt = ChatPromptTemplate.from_template("{question}")

chain = prompt | llm | StrOutputParser()

response = chain.invoke({"question": "Who won the last cricket world cup?"})
print(response)
```

### Sequential Chain

- Multiple chains connected in sequence

`SequentialChain` is used when you want to **run multiple chains step-by-step**, where:

- Output of Chain 1 ➝ Input of Chain 2
- Output of Chain 2 ➝ Input of Chain 3
- and so on…

👉 It creates a **pipeline of LLM operations**.

```python
prompt1 = ChatPromptTemplate.from_template("Give a title about {topic}")
prompt2 = ChatPromptTemplate.from_template("Write a blog about {title}")

title_chain = prompt1 | llm | StrOutputParser()
blog_chain = prompt2 | llm | StrOutputParser()

title = title_chain.invoke({"topic": "Indian Culture"})
blog = blog_chain.invoke({"title": title})

print(blog)
```
