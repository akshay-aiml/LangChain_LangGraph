```python
from typing import TypedDict
from langgraph.graph import StateGraph, END
```

**State:** This is the "shared memory." It’s a dictionary that every part of your graph can read from and write to.

**Nodes:** These are the "workers." Each node is a Python function where the actual work happens (like calling an LLM or searching the web).

**Edges:** These are the "paths." They determine which node to go to next.

Conditional Edges: These allow the AI to decide: "If the answer is 'I don't know', go back to the Search node. Otherwise, go to the End."

```python
# 1. Define the STATE (What the agent remembers)
class AgentState(TypedDict):
    messages: list[str]
```

```python
# 2. Define the NODES (The workers)
def call_model(state: AgentState):
    print("---CALLING MODEL---")
    # In a real app, you'd call an LLM here
    return {"messages": state["messages"] + ["Model processed the data!"]}


def validator(state: AgentState):
    print("---VALIDATING---")
    return state
```

```python
# 3. Build the GRAPH
workflow = StateGraph(AgentState)

# Add our workers to the map
workflow.add_node("agent", call_model)
workflow.add_node("check_quality", validator)

# Set the entry point
workflow.set_entry_point("agent")
```

```python
# 4. Define the EDGES (The logic)
workflow.add_edge("agent", "check_quality")
workflow.add_edge("check_quality", END)

```

```python
# Compile and Run
app = workflow.compile()
app.invoke({"messages": ["Hello, LangGraph!"]})
```
