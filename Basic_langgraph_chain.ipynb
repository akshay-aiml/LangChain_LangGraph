```python
from langgraph.graph import StateGraph,START,END
from langchain_groq import ChatGroq
from typing import TypedDict
```

```python
from google.colab import userdata

api_key = userdata.get("GROQ_API_KEY")
```

```python
llm = ChatGroq(model="llama-3.1-8b-instant",
               api_key=api_key,
               temperature=0.2)
```

```python
# define State
class MyState(TypedDict):
  input_topic : str
  generate_text : str
  summary : str
  keywords : list
```

```python
# define nodes to generate text
def generate_text(State:MyState) -> MyState:
  input_topic = State["input_topic"]
  prompt = f"Write a clear and informative article of approximately 300 words about the : {input_topic}"
  State["generate_text"] = llm.invoke(prompt).content
  return State
```

```python
# define node to sumarize
def summarize(State:MyState) -> MyState:
  prompt = f"Summarize this:{State['generate_text']}"
  State["summary"] = llm.invoke(prompt).content
  return State
```

```python
# define to to extract keywords
def extract_keywords(State:MyState) -> MyState:
  prompt = f"Extract keywords from this:{State['generate_text']}"
  State["keywords"] = llm.invoke(prompt).content
  return State
#
```

```python
# build the graph
builder = StateGraph(MyState)

builder.add_node("generate", generate_text)
builder.add_node("summarize", summarize)
builder.add_node("extract_keywords", extract_keywords)

builder.add_edge(START, "generate")
builder.add_edge("generate", "summarize")
builder.add_edge("summarize", "extract_keywords")
builder.add_edge("extract_keywords", END)

#compile
workflow = builder.compile()
```

```python
# test
input_topic = {"input_topic" : "Agentic AI"}

output = workflow.invoke(input_topic)
```
